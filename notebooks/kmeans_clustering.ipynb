```python
# Customer Segmentation - K-means Clustering Analysis
# --------------------------------------------------

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.decomposition import PCA
from scipy.spatial.distance import cdist
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
import joblib
import warnings
warnings.filterwarnings('ignore')

# Set visualization styles
plt.style.use('ggplot')
sns.set_palette("Set2")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("# K-means Clustering for Customer Segmentation")
print("="*50)

print("\n## 1. Loading Preprocessed Data")
print("-"*50)

# Load the preprocessed data
try:
    # Try to load PCA-transformed data first
    df = pd.read_csv('pca_prepared_data.csv')
    print("Loaded PCA-transformed data with shape:", df.shape)
    data_type = "PCA"
except FileNotFoundError:
    try:
        # If not available, try loading standard preprocessed data
        df = pd.read_csv('prepared_data.csv')
        print("Loaded standard preprocessed data with shape:", df.shape)
        data_type = "Standard"
    except FileNotFoundError:
        print("No preprocessed data found. Please run data_preprocessing.ipynb first.")
        # Create dummy data for demonstration
        print("Creating dummy data for demonstration...")
        np.random.seed(42)
        df = pd.DataFrame(
            np.random.randn(500, 10),
            columns=[f'Feature_{i+1}' for i in range(10)]
        )
        df['CustomerID'] = range(1000, 1500)
        data_type = "Dummy"

# Extract CustomerID if present
if 'CustomerID' in df.columns:
    customer_ids = df['CustomerID']
    X = df.drop(columns=['CustomerID'])
    print(f"Extracted CustomerID column. Data for clustering has shape: {X.shape}")
else:
    customer_ids = pd.Series(range(len(df)))
    X = df.copy()
    print(f"No CustomerID column found. Using data with shape: {X.shape}")

# Load preprocessing pipeline info if available
try:
    preprocessing_info = joblib.load('preprocessing_pipeline.pkl')
    print("\nPreprocessing information loaded successfully.")
    print(f"Preprocessing steps applied: {preprocessing_info['preprocessing_steps']}")
except FileNotFoundError:
    print("\nNo preprocessing information found. Proceeding with clustering directly.")
    preprocessing_info = None

print("\n## 2. Finding Optimal Number of Clusters")
print("-"*50)

# Function to evaluate clustering quality metrics
def evaluate_clusters(data, max_clusters=15):
    """Evaluate clustering quality for different numbers of clusters"""
    
    # Initialize metrics
    inertia = []
    silhouette = []
    ch_score = []
    db_score = []
    
    # Try different cluster numbers
    k_range = range(2, max_clusters + 1)
    for k in k_range:
        print(f"Evaluating k={k}...", end=" ")
        
        # Fit KMeans
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
        kmeans.fit(data)
        
        # Store inertia (within-cluster sum of squares)
        inertia.append(kmeans.inertia_)
        
        # Compute silhouette score (higher is better)
        sil = silhouette_score(data, kmeans.labels_)
        silhouette.append(sil)
        
        # Compute Calinski-Harabasz Index (higher is better)
        ch = calinski_harabasz_score(data, kmeans.labels_)
        ch_score.append(ch)
        
        # Compute Davies-Bouldin Index (lower is better)
        db = davies_bouldin_score(data, kmeans.labels_)
        db_score.append(db)
        
        print(f"Silhouette: {sil:.4f}, CH Index: {ch:.1f}, DB Index: {db:.4f}")
    
    return k_range, inertia, silhouette, ch_score, db_score

# Evaluate different numbers of clusters
max_clusters = min(15, X.shape[0] // 20)  # Don't try too many clusters for small datasets
k_range, inertia, silhouette, ch_score, db_score = evaluate_clusters(X, max_clusters)

# Plot results (Elbow Method)
plt.figure(figsize=(20, 12))

# Inertia plot (Elbow method)
plt.subplot(2, 2, 1)
plt.plot(k_range, inertia, 'o-', linewidth=2, markersize=8)
plt.title('Elbow Method', fontsize=16)
plt.xlabel('Number of clusters (k)', fontsize=14)
plt.ylabel('Inertia (Within-cluster SS)', fontsize=14)
plt.grid(True)
plt.xticks(k_range)

# Silhouette score plot (higher is better)
plt.subplot(2, 2, 2)
plt.plot(k_range, silhouette, 'o-', linewidth=2, markersize=8)
plt.title('Silhouette Score (Higher is Better)', fontsize=16)
plt.xlabel('Number of clusters (k)', fontsize=14)
plt.ylabel('Silhouette Score', fontsize=14)
plt.grid(True)
plt.xticks(k_range)

# Calinski-Harabasz Index (higher is better)
plt.subplot(2, 2, 3)
plt.plot(k_range, ch_score, 'o-', linewidth=2, markersize=8)
plt.title('Calinski-Harabasz Index (Higher is Better)', fontsize=16)
plt.xlabel('Number of clusters (k)', fontsize=14)
plt.ylabel('Calinski-Harabasz Index', fontsize=14)
plt.grid(True)
plt.xticks(k_range)

# Davies-Bouldin Index (lower is better)
plt.subplot(2, 2, 4)
plt.plot(k_range, db_score, 'o-', linewidth=2, markersize=8)
plt.title('Davies-Bouldin Index (Lower is Better)', fontsize=16)
plt.xlabel('Number of clusters (k)', fontsize=14)
plt.ylabel('Davies-Bouldin Index', fontsize=14)
plt.grid(True)
plt.xticks(k_range)

plt.tight_layout()
plt.suptitle('Determining Optimal Number of Clusters', fontsize=18, y=1.02)
plt.savefig('kmeans_optimal_clusters.png', bbox_inches='tight')
plt.show()

# Visualize elbow method using Yellowbrick
plt.figure(figsize=(10, 8))
visualizer = KElbowVisualizer(KMeans(random_state=42), k=(2, max_clusters))
visualizer.fit(X)
visualizer.show()

# Get optimal k
# Find the point of maximum curvature (elbow) using second derivative
k_diff = np.diff(inertia, 2)
elbow_idx = np.argmax(k_diff) + 2 if len(k_diff) > 0 else 2

# Also consider the maximum silhouette score
sil_idx = np.argmax(silhouette) + 2

print("\nCluster number analysis:")
print(f"Elbow method suggests {elbow_idx} clusters")
print(f"Highest silhouette score at {sil_idx} clusters (score: {max(silhouette):.4f})")

# Make recommendation on number of clusters
optimal_k = sil_idx
print(f"\nRecommended number of clusters: {optimal_k}")

# Give option to override
print("You can override this by setting 'optimal_k' to a different value")

print("\n## 3. Fitting K-means with Optimal Clusters")
print("-"*50)

# Fit K-means with optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Add cluster labels to the original data
cluster_df = pd.DataFrame({
    'CustomerID': customer_ids,
    'Cluster': y_kmeans
})

print(f"K-means model fitted with {optimal_k} clusters")
print("\nCluster distribution:")
cluster_counts = cluster_df['Cluster'].value_counts().sort_index()
for cluster_id, count in cluster_counts.items():
    percentage = count / len(cluster_df) * 100
    print(f"  Cluster {cluster_id}: {count} customers ({percentage:.1f}%)")

# Visualize final clusters using silhouette plot
plt.figure(figsize=(12, 8))
visualizer = SilhouetteVisualizer(kmeans, colors='tab10')
visualizer.fit(X)
plt.title(f'Silhouette Plot for {optimal_k} Clusters', fontsize=16)
plt.savefig('kmeans_silhouette.png', bbox_inches='tight')
plt.show()

# Save the K-means model
joblib.dump(kmeans, 'kmeans_model.pkl')
print("K-means model saved to 'kmeans_model.pkl'")

print("\n## 4. Cluster Analysis")
print("-"*50)

# Function to analyze cluster characteristics
def analyze_clusters(kmeans, data, original_data=None):
    """Analyze the characteristics of each cluster"""
    # Get cluster centers
    centers = kmeans.cluster_centers_
    labels = kmeans.labels_
    
    # Create dataframe for cluster centers
    if data_type == "PCA":
        center_df = pd.DataFrame(centers, columns=data.columns)
        print("Cluster centers in PCA space:")
        display(center_df)
        
        # We need to transform back to original space for interpretation
        # But this requires the PCA model which might not be available
        print("\nNote: PCA components need to be interpreted in terms of original features")
        print("Analyzing distribution of samples in each cluster instead")
    else:
        center_df = pd.DataFrame(centers, columns=data.columns)
        print("Cluster centers:")
        display(center_df)
    
    # If original data is available, analyze each cluster's characteristics
    if original_data is not None:
        # Add cluster labels to original data
        orig_with_clusters = original_data.copy()
        orig_with_clusters['Cluster'] = labels
        
        # Get summary statistics for each cluster
        cluster_summary = []
        
        for i in range(kmeans.n_clusters):
            cluster_data = orig_with_clusters[orig_with_clusters['Cluster'] == i]
            summary = cluster_data.describe().transpose()
            summary['cluster'] = i
            
            # Only keep mean and std
            summary = summary[['cluster', 'mean', 'std']]
            cluster_summary.append(summary)
        
        # Combine all clusters
        all_clusters_summary = pd.concat(cluster_summary)
        
        print("\nCluster summary statistics:")
        display(all_clusters_summary)
        
        return orig_with_clusters
    
    return None

# Try to load original data for better interpretation
try:
    # If we're using PCA data, try to load the original data
    if data_type == "PCA":
        original_df = pd.read_csv('customer_data.csv')
        print("Loaded original customer data for enhanced cluster interpretation")
        
        # Make sure we have the same customers in both datasets
        if 'CustomerID' in original_df.columns:
            original_df = pd.merge(
                cluster_df, original_df, 
                on='CustomerID', 
                how='inner'
            )
        else:
            # If no CustomerID in original, just use the PCA data
            original_df = None
            print("Warning: Could not link original data with clusters")
    else:
        # If we're using standard preprocessed data, we don't need original
        original_df = None
except FileNotFoundError:
    print("Original customer data not found. Using only transformed data for analysis.")
    original_df = None

# Analyze cluster characteristics
data_with_clusters = analyze_clusters(kmeans, X, original_df)

if data_with_clusters is not None:
    print("\nDetailed analysis of each cluster:")
    
    # Analyze each cluster's distinguishing characteristics
    for i in range(optimal_k):
        print(f"\nCLUSTER {i} CHARACTERISTICS:")
        print("-" * 30)
        
        # Get data for this cluster
        cluster_data = data_with_clusters[data_with_clusters['Cluster'] == i]
        
        # Compare mean values with overall mean
        if len(cluster_data) > 0:
            # Get numerical columns for comparison
            num_cols = cluster_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
            num_cols = [col for col in num_cols if col != 'Cluster']
            
            # Calculate comparison stats
            comparison = pd.DataFrame({
                'Cluster_Mean': cluster_data[num_cols].mean(),
                'Overall_Mean': data_with_clusters[num_cols].mean(),
                'Difference': cluster_data[num_cols].mean() - data_with_clusters[num_cols].mean(),
                'Percent_Diff': (cluster_data[num_cols].mean() - data_with_clusters[num_cols].mean()) / data_with_clusters[num_cols].mean() * 100
            })
            
            # Sort by absolute percentage difference
            comparison = comparison.sort_values(by='Percent_Diff', key=abs, ascending=False)
            
            # Display top differentiating features
            display(comparison.head(10))
            
            # For categorical columns, check distribution
            cat_cols = cluster_data.select_dtypes(include=['object']).columns.tolist()
            if cat_cols:
                print("\nKey categorical characteristics:")
                for col in cat_cols[:5]:  # Show top 5 categorical columns
                    print(f"\n{col} distribution:")
                    
                    # Calculate distribution in this cluster vs overall
                    cluster_dist = cluster_data[col].value_counts(normalize=True)
                    overall_dist = data_with_clusters[col].value_counts(normalize=True)
                    
                    # Create comparison dataframe
                    cat_compare = pd.DataFrame({
                        'Cluster_Pct': cluster_dist,
                        'Overall_Pct': overall_dist
                    }).fillna(0)
                    
                    cat_compare['Difference'] = cat_compare['Cluster_Pct'] - cat_compare['Overall_Pct']
                    cat_compare['Ratio'] = cat_compare['Cluster_Pct'] / cat_compare['Overall_Pct'].replace(0, 0.001)
                    
                    # Sort by difference
                    cat_compare = cat_compare.sort_values(by='Difference', ascending=False)
                    display(cat_compare.head())
            
            # Summarize key characteristics
            print("\nKey distinguishing characteristics:")
            top_features = comparison.head(5).index.tolist()
            for feature in top_features:
                diff_pct = comparison.loc[feature, 'Percent_Diff']
                direction = "higher" if diff_pct > 0 else "lower"
                print(f"- {feature}: {abs(diff_pct):.1f}% {direction} than average")
        else:
            print("No data available for this cluster.")

print("\n## 5. Visualizing Clusters")
print("-"*50)

# Function to visualize clusters
def visualize_clusters(data, labels, centers):
    """Visualize clusters using PCA for dimensionality reduction if needed"""
    
    # If data has more than 2 dimensions, reduce for visualization
    if data.shape[1] > 2:
        print("Reducing data to 2D for visualization using PCA...")
        pca_viz = PCA(n_components=2)
        data_2d = pca_viz.fit_transform(data)
        centers_2d = pca_viz.transform(centers)
        
        explained_var = pca_viz.explained_variance_ratio_
        print(f"Explained variance: {explained_var[0]:.2f}, {explained_var[1]:.2f}")
        print(f"Total variance explained: {sum(explained_var):.2f}")
    else:
        data_2d = data.values
        centers_2d = centers
    
    # Create scatter plot of samples
    plt.figure(figsize=(12, 10))
    
    # Plot points with cluster colors
    scatter = plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels, 
                         cmap='viridis', alpha=0.5, s=50)
    
    # Plot cluster centers
    plt.scatter(centers_2d[:, 0], centers_2d[:, 1], c=range(len(centers_2d)),
               cmap='viridis', s=300, marker='*', edgecolor='black', linewidth=1.5)
    
    # Add cluster labels to centers
    for i, (x, y) in enumerate(centers_2d):
        plt.annotate(f'Cluster {i}', (x, y), fontsize=14, 
                    ha='center', va='center', color='white',
                    bbox=dict(facecolor='black', alpha=0.7, boxstyle='round,pad=0.5'))
    
    # Add legend and labels
    plt.colorbar(scatter, label='Cluster')
    if data.shape[1] > 2:
        plt.xlabel(f'Principal Component 1 ({explained_var[0]:.2f} variance)', fontsize=14)
        plt.ylabel(f'Principal Component 2 ({explained_var[1]:.2f} variance)', fontsize=14)
    else:
        plt.xlabel(data.columns[0], fontsize=14)
        plt.ylabel(data.columns[1], fontsize=14)
    
    plt.title(f'K-means Clustering Results ({len(centers)} clusters)', fontsize=16)
    plt.grid(alpha=0.3)
    plt.savefig('kmeans_clusters_viz.png', bbox_inches='tight')
    plt.show()
    
    # If high-dimensional, also try TSNE visualization
    if data.shape[1] > 10 and len(data) < 5000:  # TSNE works best on smaller datasets
        print("Also visualizing with t-SNE for better cluster separation...")
        from sklearn.manifold import TSNE
        
        tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
        data_tsne = tsne.fit_transform(data)
        
        # We can't directly transform centers with TSNE, so we'll skip them
        
        plt.figure(figsize=(12, 10))
        scatter = plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=labels, 
                             cmap='viridis', alpha=0.5, s=50)
        
        plt.colorbar(scatter, label='Cluster')
        plt.xlabel('t-SNE feature 1', fontsize=14)
        plt.ylabel('t-SNE feature 2', fontsize=14)
        plt.title(f't-SNE Visualization of K-means Clusters ({len(centers)} clusters)', fontsize=16)
        plt.grid(alpha=0.3)
        plt.savefig('kmeans_tsne_viz.png', bbox_inches='tight')
        plt.show()

# Visualize the clusters
visualize_clusters(X, y_kmeans, kmeans.cluster_centers_)

# If we have original features and used PCA, try to interpret PCA components
if data_type == "PCA" and preprocessing_info is not None:
    try:
        # Load PCA model
        pca_model = joblib.load('pca_model.pkl')
        
        # Get the components
        components = pca_model.components_
        
        # Try to get original feature names
        if 'numerical_features' in preprocessing_info:
            original_features = preprocessing_info['numerical_features']
            if 'encoded_features' in preprocessing_info:
                # Add encoded categorical features
                for orig_cat, encoded_cats in preprocessing_info['encoded_features'].items():
                    original_features.extend(encoded_cats)
            
            print("\nPCA Components Interpretation:")
            print("-"*50)
            
            # Create a dataframe of component loadings
            loading_df = pd.DataFrame(
                components, 
                columns=original_features,
                index=[f'Component {i+1}' for i in range(components.shape[0])]
            )
            
            # Show top features for each component
            for i in range(min(components.shape[0], 5)):  # Show first 5 components
                print(f"\nPrincipal Component {i+1}:")
                loadings = loading_df.iloc[i].abs().sort_values(ascending=False)
                print(f"Top contributing features:")
                for feat, load in loadings.head(5).items():
                    print(f"  - {feat}: {load:.4f}")
    except FileNotFoundError:
        print("PCA model not found. Cannot interpret PCA components.")

print("\n## 6. Profiling the Customer Segments")
print("-"*50)

# Determine if we have original data for better profiling
if data_with_clusters is not None:
    # Get key distinguishing features for each cluster
    summary_data = []
    
    for i in range(optimal_k):
        cluster_data = data_with_clusters[data_with_clusters['Cluster'] == i]
        cluster_size = len(cluster_data)
        cluster_pct = cluster_size / len(data_with_clusters) * 100
        
        # Get numerical columns for comparison
        num_cols = cluster_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
        num_cols = [col for col in num_cols if col != 'Cluster']
        
        # Calculate comparison stats
        comparison = pd.DataFrame({
            'Cluster_Mean': cluster_data[num_cols].mean(),
            'Overall_Mean': data_with_clusters[num_cols].mean(),
            'Percent_Diff': (cluster_data[num_cols].mean() - data_with_clusters[num_cols].mean()) / data_with_clusters[num_cols].mean() * 100
        })
        
        # Sort by absolute percentage difference
        comparison = comparison.sort_values(by='Percent_Diff', key=abs, ascending=False)
        
        # Get top 3 distinguishing features
        top_features = comparison.head(3).index.tolist()
        key_attributes = []
        
        for feature in top_features:
            diff_pct = comparison.loc[feature, 'Percent_Diff']
            direction = "higher" if diff_pct > 0 else "lower"
            key_attributes.append(f"{feature}: {abs(diff_pct):.1f}% {direction}")
        
        summary_data.append({
            'Cluster': i,
            'Size': cluster_size,
            'Percentage': cluster_pct,
            'Key_Attributes': "; ".join(key_attributes)
        })
    
    # Create summary dataframe
    summary_df = pd.DataFrame(summary_data)
    
    print("Customer Segments Summary:")
    display(summary_df)
    
    # Now provide detailed marketing profiles for each segment
    print("\nDetailed Customer Segment Profiles:")
    print("="*80)
    
    for i in range(optimal_k):
        print(f"\n### SEGMENT {i}: {'='*40}")
        segment_data = summary_df[summary_df['Cluster'] == i].iloc[0]
        
        print(f"Size: {segment_data['Size']} customers ({segment_data['Percentage']:.1f}% of total)")
        print(f"Key distinguishing attributes: {segment_data['Key_Attributes']}")
        
        # Get data for this cluster for detailed analysis
        cluster_data = data_with_clusters[data_with_clusters['Cluster'] == i]
        
        # Look for demographic patterns
        demo_cols = [col for col in cluster_data.columns if col.lower() in 
                    ['age', 'gender', 'income', 'education', 'marital', 'occupation', 'location']]
        
        if demo_cols:
            print("\nDemographic Profile:")
            for col in demo_cols:
                if cluster_data[col].dtype in ['int64', 'float64']:
                    # Numerical demographic
                    avg = cluster_data[col].mean()
                    overall_avg = data_with_clusters[col].mean()
                    diff_pct = (avg - overall_avg) / overall_avg * 100
                    direction = "higher than" if diff_pct > 0 else "lower than"
                    print(f"- Average {col}: {avg:.1f} ({abs(diff_pct):.1f}% {direction} overall average)")
                else:
                    # Categorical demographic
                    top_cat = cluster_data[col].value_counts(normalize=True).head(1)
                    print(f"- Most common {col}: {top_cat.index[0]} ({top_cat.values[0]*100:.1f}%)")
        
        # Look for behavioral patterns
        behavior_cols = [col for col in cluster_data.columns if col.lower() in 
                        ['spend', 'purchase', 'visit', 'frequency', 'recency', 'loyalty']]
        
        if behavior_cols:
            print("\nBehavioral Profile:")
            for col in behavior_cols:
                if cluster_data[col].dtype in ['int64', 'float64']:
                    # Numerical behavioral metric
                    avg = cluster_data[col].mean()
                    overall_avg = data_with_clusters[col].mean()
                    diff_pct = (avg - overall_avg) / overall_avg * 100
                    direction = "higher than" if diff_pct > 0 else "lower than"
                    print(f"- Average {col}: {avg:.1f} ({abs(diff_pct):.1f}% {direction} overall average)")
        
        # Provide suggested marketing approach
        print("\nMarketing Recommendations:")
        
        # Very simplified approach to recommendations based on segment characteristics
        # In real life, this would be much more nuanced and based on business domain knowledge
        if 'Income' in comparison.index or 'Annual_Income' in comparison.index:
            income_feature = 'Income' if 'Income' in comparison.index else 'Annual_Income'
            income_diff = comparison.loc[income_feature, 'Percent_Diff']
            
            if income_diff > 20:
                print("- Premium positioning: Target with high-end products and exclusive offers")
            elif income_diff < -20:
                print("- Value positioning: Emphasize affordability and value for money")
        
        if 'Age' in comparison.index:
            age_diff = comparison.loc['Age', 'Percent_Diff']
            
            if age_diff > 20:
                print("- Focus on traditional media channels and products for older demographics")
            elif age_diff < -20:
                print("- Prioritize digital channels and trendy products for younger audiences")
        
        if any(col for col in comparison.index if 'Spend' in col or 'Purchase' in col):
            spend_cols = [col for col in comparison.index if 'Spend' in col or 'Purchase' in col]
            avg_spend_diff = sum(comparison.loc[col, 'Percent_Diff'] for col in spend_cols) / len(spend_cols)
            
            if avg_spend_diff > 20:
                print("- Develop loyalty programs to retain these high-value customers")
                print("- Cross-sell and upsell premium products and services")
            elif avg_spend_diff < -20:
                print("- Create entry-level offers to increase spending")
                print("- Implement win-back campaigns with special promotions")
        
        print("\n" + "-"*80)
    
    # Save segmented data for future use
    data_with_clusters.to_csv('kmeans_segmented_customers.csv', index=False)
    print("\nSegmented customer data saved to 'kmeans_segmented_customers.csv'")
else:
    print("Unable to generate detailed profiles without original customer data")
    
    # Just save the cluster assignments
    cluster_df.to_csv('kmeans_cluster_assignments.csv', index=False)
    print("\nCluster assignments saved to 'kmeans_cluster_assignments.csv'")

print("\n## 7. Summary")
print("-"*50)

print(f"1. Performed K-means clustering with {optimal_k} clusters")
print(f"2. Model silhouette score: {silhouette_score(X, y_kmeans):.4f}")
print(f"3. Cluster sizes vary from {cluster_df['Cluster'].value_counts().min()} to {cluster_df['Cluster'].value_counts().max()} customers")

print("\nNext Steps:")
print("1. Use the segmentation for targeted marketing campaigns")
print("2. Compare with DBSCAN clustering results")
print("3. Validate segments with business stakeholders")
print("4. Implement segment-specific strategies")

# Create a concise version of the results for quick reference
summary_stats = {
    'algorithm': 'K-means',
    'num_clusters': optimal_k,
    'silhouette_score': silhouette_score(X, y_kmeans),
    'calinski_harabasz_score': calinski_harabasz_score(X, y_kmeans),
    'davies_bouldin_score': davies_bouldin_score(X, y_kmeans),
    'cluster_sizes': cluster_df['Cluster'].value_counts().to_dict()
}

# Save summary statistics
joblib.dump(summary_stats, 'kmeans_summary_stats.pkl')
print("\nSummary statistics saved to 'kmeans_summary_stats.pkl'")

print("\nK-means clustering analysis complete!")
```
