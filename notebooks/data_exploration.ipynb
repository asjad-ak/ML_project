```python
# Customer Segmentation - Data Exploration
# ----------------------------------------

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import missingno as msno
import warnings
warnings.filterwarnings('ignore')

# Set visualization styles
plt.style.use('ggplot')
sns.set_palette("Set2")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# Load the customer data
print("Loading customer dataset...")
# Assuming your dataset is in the current directory
# Change the path if it's located elsewhere
try:
    df = pd.read_csv('customer_data.csv')
    print(f"Dataset loaded successfully with {df.shape[0]} rows and {df.shape[1]} columns.")
except FileNotFoundError:
    print("Dataset file not found. Please ensure your dataset is named 'customer_data.csv' or adjust the path.")
    # Creating a sample dataset for demonstration
    print("Creating a sample dataset for demonstration...")
    np.random.seed(42)
    df = pd.DataFrame({
        'CustomerID': range(1000, 1500),
        'Age': np.random.randint(18, 80, 500),
        'Annual_Income': np.random.normal(45000, 15000, 500),
        'Spending_Score': np.random.randint(1, 100, 500),
        'Years_as_Customer': np.random.randint(0, 15, 500),
        'Purchase_Frequency': np.random.normal(10, 5, 500),
        'Gender': np.random.choice(['Male', 'Female'], 500),
        'Education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 500),
        'Marital_Status': np.random.choice(['Single', 'Married', 'Divorced'], 500),
    })

# Display basic dataset information
print("\n1. Basic Dataset Information:")
print("-" * 30)
print(df.info())

# Display first few rows of the dataset
print("\n2. First 5 rows of the dataset:")
print("-" * 30)
display(df.head())

# Statistical summary of numerical features
print("\n3. Statistical Summary of Numerical Features:")
print("-" * 30)
display(df.describe())

# Check for missing values
print("\n4. Missing Values Analysis:")
print("-" * 30)
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_data = pd.concat([missing_values, missing_percent], axis=1)
missing_data.columns = ['Missing Values', 'Percentage (%)']
missing_data = missing_data[missing_data['Missing Values'] > 0]
print(missing_data if not missing_data.empty else "No missing values found!")

# Visualize missing values
plt.figure(figsize=(12, 6))
msno.matrix(df)
plt.title('Missing Values Matrix', fontsize=16)
plt.show()

# Explore categorical variables
print("\n5. Categorical Variables Distribution:")
print("-" * 30)
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
if categorical_cols:
    for col in categorical_cols:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=col, data=df)
        plt.title(f'Distribution of {col}', fontsize=16)
        plt.xticks(rotation=45)
        plt.show()
        
        print(f"\nFrequency Distribution for {col}:")
        display(df[col].value_counts().reset_index())
        print(f"Percentage Distribution for {col}:")
        display(df[col].value_counts(normalize=True).reset_index().rename(
            columns={'index': col, col: 'Percentage'}))
        print("-" * 50)
else:
    print("No categorical variables found in the dataset.")

# Explore numerical variables
print("\n6. Numerical Variables Distribution:")
print("-" * 30)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Histograms for numerical features
for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}', fontsize=14)
    
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}', fontsize=14)
    plt.tight_layout()
    plt.show()

# Correlation analysis
print("\n7. Correlation Analysis:")
print("-" * 30)
if len(numerical_cols) > 1:
    corr_matrix = df[numerical_cols].corr()
    
    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    
    # Heatmap visualization
    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=True, mask=mask, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Correlation Matrix of Numerical Features', fontsize=16)
    plt.show()
    
    # Find strong correlations
    print("\nStrong Correlations (|corr| > 0.5):")
    strong_corr = pd.DataFrame()
    for i, col1 in enumerate(corr_matrix.columns):
        for col2 in corr_matrix.columns[i+1:]:
            if abs(corr_matrix.loc[col1, col2]) > 0.5:
                strong_corr = strong_corr.append({
                    'Feature 1': col1,
                    'Feature 2': col2,
                    'Correlation': corr_matrix.loc[col1, col2]
                }, ignore_index=True)
    
    if not strong_corr.empty:
        display(strong_corr.sort_values('Correlation', key=abs, ascending=False))
    else:
        print("No strong correlations found.")
else:
    print("Not enough numerical features for correlation analysis.")

# Exploring relationships between numerical and categorical variables
print("\n8. Relationship Between Numerical and Categorical Variables:")
print("-" * 30)
if categorical_cols and numerical_cols:
    for cat_col in categorical_cols[:2]:  # Limiting to 2 categorical features to avoid too many plots
        for num_col in numerical_cols[:3]:  # Limiting to 3 numerical features
            plt.figure(figsize=(12, 6))
            
            # Box plot
            plt.subplot(1, 2, 1)
            sns.boxplot(x=cat_col, y=num_col, data=df)
            plt.title(f'{num_col} by {cat_col} (Boxplot)', fontsize=14)
            plt.xticks(rotation=45)
            
            # Violin plot
            plt.subplot(1, 2, 2)
            sns.violinplot(x=cat_col, y=num_col, data=df)
            plt.title(f'{num_col} by {cat_col} (Violin Plot)', fontsize=14)
            plt.xticks(rotation=45)
            
            plt.tight_layout()
            plt.show()
            
            # Calculate group statistics
            print(f"\nStatistics of {num_col} grouped by {cat_col}:")
            display(df.groupby(cat_col)[num_col].describe())
            print("-" * 50)
else:
    print("Either categorical or numerical variables are missing for this analysis.")

# Pair plot for main numerical features (limited to avoid overplotting)
print("\n9. Pair Plot for Key Numerical Features:")
print("-" * 30)
if len(numerical_cols) > 1:
    key_features = numerical_cols[:4]  # Limit to 4 features to keep visualization readable
    plt.figure(figsize=(16, 12))
    sns.pairplot(df[key_features + categorical_cols[:1]], hue=categorical_cols[0] if categorical_cols else None)
    plt.suptitle('Pair Plot of Key Features', y=1.02, fontsize=16)
    plt.show()
else:
    print("Not enough numerical features for pair plot analysis.")

# Feature outlier analysis
print("\n10. Outlier Analysis:")
print("-" * 30)
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_percentage = len(outliers) / len(df) * 100
    
    print(f"Feature: {col}")
    print(f"Number of outliers: {len(outliers)}")
    print(f"Percentage of outliers: {outlier_percentage:.2f}%")
    print(f"Outlier threshold: < {lower_bound:.2f} or > {upper_bound:.2f}")
    print("-" * 50)

# Summary of key findings
print("\n11. Summary of Key Findings:")
print("-" * 30)
print("Based on the exploratory data analysis:")
print("1. Dataset shape: {0} rows and {1} columns".format(df.shape[0], df.shape[1]))
print("2. Missing values: " + ("Present" if df.isnull().sum().sum() > 0 else "None detected"))

# Find the column with highest variance (normalized)
if numerical_cols:
    cv_values = df[numerical_cols].std() / df[numerical_cols].mean()
    highest_var_feature = cv_values.idxmax()
    print(f"3. Feature with highest relative variability: {highest_var_feature}")

# Identify potential segmentation variables
print("4. Potential customer segmentation features to consider:")
print("   - Demographic features: " + ", ".join([col for col in df.columns if col in ['Age', 'Gender', 'Income', 'Education', 'Marital_Status']]))
print("   - Behavioral features: " + ", ".join([col for col in df.columns if col in ['Spending_Score', 'Purchase_Frequency', 'Years_as_Customer']]))

print("\n12. Next Steps for Clustering Analysis:")
print("-" * 30)
print("1. Handle missing values (if any)")
print("2. Encode categorical variables")
print("3. Scale/normalize numerical features")
print("4. Apply dimensionality reduction techniques if needed")
print("5. Run K-Means and DBSCAN clustering algorithms")
print("6. Evaluate cluster quality and interpret results")

# Add a function to help with initial feature selection
def suggest_features(df):
    """Suggest features for clustering based on variance and correlation."""
    print("\n13. Feature Selection Recommendations:")
    print("-" * 30)
    
    # Calculate coefficient of variation for numerical features
    if numerical_cols:
        cv = pd.DataFrame({
            'Feature': numerical_cols,
            'CV': [df[col].std() / abs(df[col].mean()) if df[col].mean() != 0 else df[col].std() for col in numerical_cols]
        }).sort_values('CV', ascending=False)
        
        print("Features ranked by variability (higher is better for clustering):")
        display(cv)
        
        # Suggest features with higher variation
        high_var_features = cv[cv['CV'] > cv['CV'].median()]['Feature'].tolist()
        print(f"\nRecommended numerical features with high variability: {', '.join(high_var_features)}")
    
    # For categorical features, look at entropy
    if categorical_cols:
        print("\nCategorical features by number of unique values:")
        cat_unique = pd.DataFrame({
            'Feature': categorical_cols,
            'Unique_Values': [df[col].nunique() for col in categorical_cols]
        }).sort_values('Unique_Values', ascending=False)
        display(cat_unique)

suggest_features(df)

print("\nExploratory Data Analysis completed.")
```
