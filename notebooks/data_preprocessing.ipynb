```python
# Customer Segmentation - Data Preprocessing
# ------------------------------------------

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import joblib
import warnings
warnings.filterwarnings('ignore')

# Set visualization styles
plt.style.use('ggplot')
sns.set_palette("Set2")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("# Customer Data Preprocessing Pipeline")
print("="*50)

# Load the customer data
print("\n## 1. Loading Dataset")
print("-"*50)

try:
    df = pd.read_csv('customer_data.csv')
    print(f"Dataset loaded successfully with {df.shape[0]} rows and {df.shape[1]} columns.")
except FileNotFoundError:
    print("Dataset file not found. Please ensure your dataset is named 'customer_data.csv' or adjust the path.")
    # Creating a sample dataset for demonstration
    print("Creating a sample dataset for demonstration...")
    np.random.seed(42)
    df = pd.DataFrame({
        'CustomerID': range(1000, 1500),
        'Age': np.random.randint(18, 80, 500),
        'Annual_Income': np.random.normal(45000, 15000, 500),
        'Spending_Score': np.random.randint(1, 100, 500),
        'Years_as_Customer': np.random.randint(0, 15, 500),
        'Purchase_Frequency': np.random.normal(10, 5, 500),
        'Gender': np.random.choice(['Male', 'Female'], 500),
        'Education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 500),
        'Marital_Status': np.random.choice(['Single', 'Married', 'Divorced'], 500),
    })

# Make a copy of original data
original_df = df.copy()

# Display first few rows of the dataset
print("\nFirst 5 rows of the dataset:")
display(df.head())

# Data types and information
print("\nDataset information:")
print(df.info())

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"\nNumerical columns ({len(numerical_cols)}): {', '.join(numerical_cols)}")
print(f"Categorical columns ({len(categorical_cols)}): {', '.join(categorical_cols)}")

# Function to identify ID columns that shouldn't be used for clustering
def identify_id_columns(df):
    potential_id_cols = []
    for col in df.columns:
        # Check if column name contains ID-related terms
        if 'id' in col.lower() or 'key' in col.lower() or 'code' in col.lower():
            # Check if values are unique or nearly unique (>90% unique values)
            uniqueness = df[col].nunique() / len(df)
            if uniqueness > 0.9:
                potential_id_cols.append(col)
    return potential_id_cols

id_columns = identify_id_columns(df)
print(f"\nPotential ID columns to exclude from clustering: {', '.join(id_columns)}")

# Remove ID columns for clustering analysis
if id_columns:
    print(f"\nRemoving ID columns: {', '.join(id_columns)}")
    df_no_id = df.drop(columns=id_columns)
    # Store customer IDs for later reference
    customer_ids = df[id_columns[0]] if id_columns else pd.Series(range(len(df)))
else:
    df_no_id = df.copy()
    customer_ids = pd.Series(range(len(df)))

print("\n## 2. Handling Missing Values")
print("-"*50)

# Check for missing values
missing_values = df_no_id.isnull().sum()
missing_percent = (missing_values / len(df_no_id)) * 100
missing_data = pd.concat([missing_values, missing_percent], axis=1)
missing_data.columns = ['Missing Values', 'Percentage (%)']
missing_data = missing_data[missing_data['Missing Values'] > 0]

if not missing_data.empty:
    print("\nMissing values in the dataset:")
    display(missing_data)
    
    # Visualize missing values
    plt.figure(figsize=(12, 6))
    sns.heatmap(df_no_id.isnull(), cmap='viridis', cbar=False, yticklabels=False)
    plt.title('Missing Values Heatmap', fontsize=16)
    plt.tight_layout()
    plt.show()

    # Function to handle missing values based on data type and missing percentage
    def handle_missing_values(df, col, strategy='auto'):
        missing_pct = df[col].isnull().mean() * 100
        
        # If less than 5% missing, use mean/mode imputation
        if missing_pct < 5 or strategy == 'simple':
            if df[col].dtype in ['int64', 'float64']:
                print(f"Imputing missing values in '{col}' with mean (simple imputation)")
                imputer = SimpleImputer(strategy='mean')
                df[col] = imputer.fit_transform(df[[col]])
            else:
                print(f"Imputing missing values in '{col}' with most frequent value")
                imputer = SimpleImputer(strategy='most_frequent')
                df[col] = imputer.fit_transform(df[[col]])
        
        # If between 5% and 20% missing, use KNN imputation for numerical features
        elif 5 <= missing_pct < 20 or strategy == 'knn':
            if df[col].dtype in ['int64', 'float64']:
                print(f"Imputing missing values in '{col}' with KNN imputation")
                # Only use numerical features for KNN imputation
                num_df = df[numerical_cols].copy()
                imputer = KNNImputer(n_neighbors=5)
                imputed_values = imputer.fit_transform(num_df)
                col_idx = numerical_cols.index(col)
                df[col] = imputed_values[:, col_idx]
            else:
                print(f"Imputing missing values in '{col}' with most frequent value")
                imputer = SimpleImputer(strategy='most_frequent')
                df[col] = imputer.fit_transform(df[[col]])
        
        # If more than 20% missing, either drop column or create a "Missing" category
        elif missing_pct >= 20:
            if df[col].dtype in ['int64', 'float64']:
                print(f"Warning: High missing rate ({missing_pct:.1f}%) in '{col}'. Using median imputation.")
                imputer = SimpleImputer(strategy='median')
                df[col] = imputer.fit_transform(df[[col]])
                
                # Create a binary flag for missingness
                df[f'{col}_was_missing'] = df[col].isnull().astype(int)
                print(f"Created missingness indicator: '{col}_was_missing'")
            else:
                print(f"Creating 'Missing' category for '{col}' with {missing_pct:.1f}% missing values")
                df[col] = df[col].fillna('Missing')
    
    # Handle missing values for each column with missing data
    for col in missing_data.index:
        handle_missing_values(df_no_id, col)
    
    print("\nMissing values after imputation:")
    display(df_no_id.isnull().sum())
else:
    print("No missing values found in the dataset.")

print("\n## 3. Handling Outliers")
print("-"*50)

# Function to detect and handle outliers
def handle_outliers(df, method='iqr', columns=None):
    if columns is None:
        columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    outlier_summary = []
    
    for col in columns:
        if method == 'iqr':
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Count outliers
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            outlier_count = len(outliers)
            outlier_percent = outlier_count / len(df) * 100
            
            outlier_summary.append({
                'Feature': col,
                'Outliers': outlier_count,
                'Percentage': outlier_percent,
                'Lower Bound': lower_bound,
                'Upper Bound': upper_bound
            })
            
            # Cap outliers
            if outlier_count > 0:
                df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
                print(f"Capped {outlier_count} outliers ({outlier_percent:.2f}%) in '{col}'")
        
        elif method == 'zscore':
            z_scores = (df[col] - df[col].mean()) / df[col].std()
            outliers = df[abs(z_scores) > 3]
            outlier_count = len(outliers)
            outlier_percent = outlier_count / len(df) * 100
            
            outlier_summary.append({
                'Feature': col,
                'Outliers': outlier_count,
                'Percentage': outlier_percent,
                'Z-Score Threshold': 3
            })
            
            # Cap outliers
            if outlier_count > 0:
                df[col] = df[col].mask(abs(z_scores) > 3, df[col].mean())
                print(f"Replaced {outlier_count} outliers ({outlier_percent:.2f}%) with mean in '{col}'")
    
    return pd.DataFrame(outlier_summary)

# Handle outliers using IQR method
outlier_summary = handle_outliers(df_no_id, method='iqr')
print("\nOutlier Summary:")
display(outlier_summary)

# Visualize numerical distributions before and after outlier treatment
for col in numerical_cols[:4]:  # Limiting to first 4 numerical columns for brevity
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    sns.histplot(original_df[col], kde=True, color='blue', label='Original')
    plt.title(f'Original Distribution of {col}', fontsize=14)
    
    plt.subplot(1, 2, 2)
    sns.histplot(df_no_id[col], kde=True, color='green', label='After Treatment')
    plt.title(f'Distribution After Outlier Treatment - {col}', fontsize=14)
    
    plt.tight_layout()
    plt.show()

print("\n## 4. Feature Engineering")
print("-"*50)

# Function to create new features
def create_features(df):
    feature_df = df.copy()
    features_created = []
    
    # Example feature engineering (adjust based on your actual data)
    # Create age groups if Age column exists
    if 'Age' in feature_df.columns:
        feature_df['Age_Group'] = pd.cut(
            feature_df['Age'], 
            bins=[0, 25, 35, 45, 55, 65, 100],
            labels=['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
        )
        features_created.append('Age_Group')
    
    # Create income groups if Annual_Income column exists
    if 'Annual_Income' in feature_df.columns:
        feature_df['Income_Level'] = pd.qcut(
            feature_df['Annual_Income'],
            q=4,
            labels=['Low', 'Medium-Low', 'Medium-High', 'High']
        )
        features_created.append('Income_Level')
    
    # Create spending categories if Spending_Score exists
    if 'Spending_Score' in feature_df.columns:
        feature_df['Spending_Category'] = pd.qcut(
            feature_df['Spending_Score'],
            q=3,
            labels=['Low', 'Medium', 'High']
        )
        features_created.append('Spending_Category')
    
    # Customer value ratio (if both income and spending exist)
    if 'Annual_Income' in feature_df.columns and 'Spending_Score' in feature_df.columns:
        feature_df['Value_Ratio'] = feature_df['Spending_Score'] / (feature_df['Annual_Income'] / 10000)
        features_created.append('Value_Ratio')
    
    # Customer loyalty index (if appropriate columns exist)
    if 'Years_as_Customer' in feature_df.columns and 'Purchase_Frequency' in feature_df.columns:
        feature_df['Loyalty_Index'] = feature_df['Years_as_Customer'] * feature_df['Purchase_Frequency']
        features_created.append('Loyalty_Index')
    
    print(f"Created {len(features_created)} new features: {', '.join(features_created)}")
    return feature_df, features_created

# Create new features
df_featured, new_features = create_features(df_no_id)

# Display the enhanced dataset
print("\nDataset with new features:")
display(df_featured.head())

# Update categorical and numerical columns lists
categorical_cols = df_featured.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_cols = df_featured.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("\n## 5. Feature Encoding")
print("-"*50)

# Function to encode categorical features
def encode_categorical_features(df, encoding_method='onehot'):
    df_encoded = df.copy()
    encoded_features = {}
    encoders = {}
    
    for col in categorical_cols:
        if encoding_method == 'onehot':
            # For features with few categories, use one-hot encoding
            if df_encoded[col].nunique() < 10:
                print(f"One-hot encoding '{col}' with {df_encoded[col].nunique()} categories")
                encoder = OneHotEncoder(sparse_output=False, drop='first')
                encoded_data = encoder.fit_transform(df_encoded[[col]])
                
                # Create encoded column names
                encoded_col_names = [f"{col}_{cat}" for cat in encoder.categories_[0][1:]]
                
                # Add encoded columns to dataframe
                for i, encoded_col in enumerate(encoded_col_names):
                    df_encoded[encoded_col] = encoded_data[:, i]
                
                # Store original and encoded feature names
                encoded_features[col] = encoded_col_names
                encoders[col] = encoder
                
                # Drop original column
                df_encoded.drop(columns=[col], inplace=True)
        
        elif encoding_method == 'label':
            # Use label encoding for ordinal features or features with many categories
            print(f"Label encoding '{col}' with {df_encoded[col].nunique()} categories")
            encoder = LabelEncoder()
            df_encoded[f"{col}_encoded"] = encoder.fit_transform(df_encoded[col])
            
            # Store mapping for reference
            mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
            print(f"  Mapping: {mapping}")
            
            # Store encoder and mapping
            encoders[col] = encoder
            encoded_features[col] = [f"{col}_encoded"]
            
            # Drop original column
            df_encoded.drop(columns=[col], inplace=True)
    
    # Save encoders for future use
    joblib.dump(encoders, 'categorical_encoders.pkl')
    print(f"\nSaved encoders to 'categorical_encoders.pkl'")
    
    return df_encoded, encoded_features

# Encode categorical features
df_encoded, encoded_features = encode_categorical_features(df_featured, encoding_method='onehot')

print("\nDataset after encoding categorical features:")
display(df_encoded.head())

print("\n## 6. Feature Scaling")
print("-"*50)

# Function to scale numerical features
def scale_features(df, method='standard'):
    df_scaled = df.copy()
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    if method == 'standard':
        print(f"Applying StandardScaler to {len(num_cols)} numerical features")
        scaler = StandardScaler()
    elif method == 'minmax':
        print(f"Applying MinMaxScaler to {len(num_cols)} numerical features")
        scaler = MinMaxScaler()
    elif method == 'robust':
        print(f"Applying RobustScaler to {len(num_cols)} numerical features")
        scaler = RobustScaler()
    
    if num_cols:
        scaled_data = scaler.fit_transform(df_scaled[num_cols])
        
        # Replace original columns with scaled values
        for i, col in enumerate(num_cols):
            df_scaled[col] = scaled_data[:, i]
        
        # Save scaler for future use
        joblib.dump(scaler, f'{method}_scaler.pkl')
        print(f"Saved {method} scaler to '{method}_scaler.pkl'")
    
    return df_scaled, scaler

# Scale features using StandardScaler
df_scaled, scaler = scale_features(df_encoded, method='standard')

print("\nDataset after scaling numerical features:")
display(df_scaled.head())

print("\n## 7. Dimensionality Reduction (Optional)")
print("-"*50)

# Decide whether to perform dimensionality reduction
num_features = df_scaled.shape[1]
perform_pca = num_features > 10

# Add this before PCA section
print("\n## Ensuring all features are numeric before dimensionality reduction")
print("-"*50)

# Check for any non-numeric columns that might have been missed
non_numeric_cols = df_scaled.select_dtypes(exclude=['int64', 'float64']).columns.tolist()

if non_numeric_cols:
    print(f"Found {len(non_numeric_cols)} non-numeric columns that need handling: {non_numeric_cols}")
    
    for col in non_numeric_cols:
        print(f"Converting or dropping column: {col}")
        # Check unique values
        unique_vals = df_scaled[col].nunique()
        print(f"  - Contains {unique_vals} unique values")
        
        if unique_vals < 10:  # If few categories, one-hot encode
            print(f"  - One-hot encoding '{col}'")
            # Create dummies and drop the first to avoid multicollinearity
            dummies = pd.get_dummies(df_scaled[col], prefix=col, drop_first=True)
            df_scaled = pd.concat([df_scaled, dummies], axis=1)
        else:
            print(f"  - Too many categories, dropping column '{col}'")
        
        # Drop the original column
        df_scaled = df_scaled.drop(columns=[col])
    
    print(f"After handling non-numeric columns, shape: {df_scaled.shape}")
else:
    print("All columns are numeric, proceeding with PCA.")

if perform_pca:
    # Apply PCA for dimensionality reduction
    print(f"Applying PCA for dimensionality reduction (dataset has {num_features} features)")
    
    # Determine number of components to retain 95% of variance
    pca = PCA(n_components=0.95)
    pca_result = pca.fit_transform(df_scaled)
    
    print(f"Reduced dimensions from {num_features} to {pca.n_components_} components")
    print(f"Explained variance ratios: {pca.explained_variance_ratio_}")
    
    # Create DataFrame with PCA results
    pca_df = pd.DataFrame(
        data=pca_result,
        columns=[f'PC{i+1}' for i in range(pca.n_components_)]
    )
    
    # Visualize explained variance
    plt.figure(figsize=(10, 6))
    plt.bar(
        range(1, len(pca.explained_variance_ratio_) + 1),
        pca.explained_variance_ratio_,
        alpha=0.8,
        color='skyblue'
    )
    plt.step(
        range(1, len(pca.explained_variance_ratio_) + 1),
        np.cumsum(pca.explained_variance_ratio_),
        where='mid',
        color='red',
        marker='o'
    )
    plt.axhline(y=0.95, color='k', linestyle='--', alpha=0.5)
    plt.title('Explained Variance by Components', fontsize=16)
    plt.xlabel('Number of Components', fontsize=14)
    plt.ylabel('Explained Variance Ratio', fontsize=14)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Save PCA model
    joblib.dump(pca, 'pca_model.pkl')
    print("Saved PCA model to 'pca_model.pkl'")
    
    # Add customer IDs to PCA DataFrame for reference
    pca_df['CustomerID'] = customer_ids.values
    
    # Save prepared data for modeling
    pca_df.to_csv('pca_prepared_data.csv', index=False)
    print("Saved PCA-transformed data to 'pca_prepared_data.csv'")
    
    # Preview the PCA-transformed data
    print("\nPCA-transformed data (first 5 rows):")
    display(pca_df.head())
    
    # Final dataset for modeling will be the PCA result
    final_df = pca_df
else:
    print(f"Skipping dimensionality reduction (dataset has {num_features} features)")
    
    # Add customer IDs to scaled DataFrame for reference
    df_scaled['CustomerID'] = customer_ids.values
    
    # Save prepared data for modeling
    df_scaled.to_csv('prepared_data.csv', index=False)
    print("Saved prepared data to 'prepared_data.csv'")
    
    # Final dataset for modeling will be the scaled data
    final_df = df_scaled

print("\n## 8. Summary of Preprocessing Steps")
print("-"*50)
print(f"1. Loaded dataset with {original_df.shape[0]} rows and {original_df.shape[1]} columns")
print(f"2. Identified and handled missing values")
print(f"3. Detected and handled outliers in numerical features")
print(f"4. Created {len(new_features)} new features: {', '.join(new_features)}")
print(f"5. Encoded categorical features")
print(f"6. Scaled numerical features using StandardScaler")
if perform_pca:
    print(f"7. Applied PCA dimensionality reduction (from {num_features} to {pca.n_components_} components)")

print(f"\nFinal dataset has {final_df.shape[0]} rows and {final_df.shape[1]} columns")
print(f"Data saved and ready for clustering algorithms")

# Create a pipeline summary for the preprocessing steps
preprocessing_steps = []
preprocessing_steps.append("Handle missing values")
preprocessing_steps.append("Handle outliers")
preprocessing_steps.append("Feature engineering")
preprocessing_steps.append("Categorical encoding")
preprocessing_steps.append("Feature scaling")
if perform_pca:
    preprocessing_steps.append("PCA dimensionality reduction")

preprocessing_pipeline = {
    'preprocessing_steps': preprocessing_steps,
    'categorical_features': categorical_cols,
    'numerical_features': numerical_cols,
    'engineered_features': new_features,
    'encoded_features': encoded_features,
    'scaling_method': 'standard',
    'dim_reduction': 'PCA' if perform_pca else None
}

# Save preprocessing pipeline information
joblib.dump(preprocessing_pipeline, 'preprocessing_pipeline.pkl')
print("\nSaved preprocessing pipeline information to 'preprocessing_pipeline.pkl'")

print("\nData preprocessing completed successfully!")
```
